```python
import numpy as np
import pandas as pd
import tkinter as tk
import tkinter.font as font

red = pd.read_csv('Red.csv')
white = pd.read_csv('White.csv')
sparkling = pd.read_csv('Sparkling.csv')
rose = pd.read_csv('Rose.csv')
red['WineStyle'] = 'red'
white['WineStyle'] = 'white'
sparkling['WineStyle'] = 'sparkling'
rose['WineStyle'] = 'rose'
wines = pd.concat([red, white, sparkling, rose], ignore_index=True)
wines['Year'] = wines['Year'].replace('N.V.', 2021)
wines['Year'] = wines['Year'].astype('int')
wines['Year'] = abs(wines['Year']-2022)
wines['coefic'] = wines['Rating'] / wines['Price']
wines.loc[wines['NumberOfRatings'] < 100, ['NumberOfRatings']] = 2
wines.loc[wines['NumberOfRatings'] >=1000, ['NumberOfRatings']] = 4
wines.loc[wines['NumberOfRatings'] >=100, ['NumberOfRatings']] = 3
wines['NumberOfRatings'] = wines['NumberOfRatings'].replace(2, 'Unpopular')
wines['NumberOfRatings'] = wines['NumberOfRatings'].replace(3, 'Semi-popular')
wines['NumberOfRatings'] = wines['NumberOfRatings'].replace(4, 'Popular')
wines['NumberOfRatings'] = wines['NumberOfRatings'].astype('str')

def zatwierdz(country, style, popularity, price, agedown, ageup, wines, stronka):
    def f():
        f = font.Font(family='Times New Roman', size=7, weight="bold")
        kraj = country.get()
        styl = style.get()
        czestosc = popularity.get()
        cena = price.get()
        cena = float(cena)
        wiekmin = agedown.get()
        wiekmax = ageup.get()
        if kraj != 'none':
            wines1 = wines.loc[wines['Country'] == kraj]
        else:
            wines1 = wines
        if styl != 'none':
            wines2 = wines1.loc[wines1['WineStyle'] == styl]
        else:
            wines2 = wines1
        if czestosc != 'none':
            wines3 = wines2.loc[wines2['NumberOfRatings'] == czestosc]
        else:
            wines3 = wines2
        wines4 = wines3.loc[wines3['Price'] <= cena]
        wines5 = wines4.loc[wines4['Year'] >= wiekmin]
        wines6 = wines5.loc[wines5['Year'] <= wiekmax]
        posrt = wines6.sort_values(by=['Rating', 'Price'], ascending=(False, True))
        posrt =pd.concat([posrt], ignore_index=True)
        stronka.destroy()
        stronka1 = tk.Tk()
        stronka1.geometry("1500x1500")
        stronka1.title("Wines for you")
        gorny_panel = tk.Label(stronka1, bg="white", fg="red", width=200,
                               text="That's your wines",
                               font=18, anchor="center")
        gorny_panel.pack()
        if len(posrt['Name']) >= 1:
            top = tk.Label(stronka1, bg="white", fg="black", width=200,
                                text='1: ' + posrt['Name'][0] + ' in price '+ str(posrt['Price'][0])+ ' Euro, with medium rating '+ str(posrt['Rating'][0]) + ' from '+ str(posrt['Winery'][0]) + ' winery',
                                font=10, anchor="center")
            top['font'] = f
            top.pack()
        if len(posrt['Name']) >= 2:
            top1 = tk.Label(stronka1, bg="white", fg="black", width=200,
                           text='2: ' + posrt['Name'][1] + ' in price '+ str(posrt['Price'][1])+ ' Euro, with medium rating '+ str(posrt['Rating'][1]) + ' from '+ str(posrt['Winery'][1]) + ' winery',
                           font=10, anchor="center")
            top1['font'] = f
            top1.pack()
        if len(posrt['Name']) >= 3:
            top2 = tk.Label(stronka1, bg="white", fg="black", width=200,
                           text='3: ' + posrt['Name'][2] + ' in price '+ str(posrt['Price'][2])+ ' Euro, with medium rating '+ str(posrt['Rating'][2]) + ' from '+ str(posrt['Winery'][2]) + ' winery',
                           font=10, anchor="center")
            top2['font'] = f
            top2.pack()
        if len(posrt['Name']) >= 4:
            top3 = tk.Label(stronka1, bg="white", fg="black", width=200,
                           text='4: ' + posrt['Name'][3] + ' in price '+ str(posrt['Price'][3])+ ' Euro, with medium rating '+ str(posrt['Rating'][3]) + ' from '+ str(posrt['Winery'][3]) + ' winery',
                           font=10, anchor="center")
            top3['font'] = f
            top3.pack()
        if len(posrt['Name']) >= 5:
            top4 = tk.Label(stronka1, bg="white", fg="black", width=200,
                           text='5: ' + posrt['Name'][4] + ' in price '+ str(posrt['Price'][4])+ ' Euro, with medium rating '+ str(posrt['Rating'][4]) + ' from '+ str(posrt['Winery'][4]) + ' winery',
                           font=10, anchor="center")
            top4['font'] = f
            top4.pack()
        if len(posrt['Name']) >= 6:
            posrt2 = wines6.sort_values(by=['coefic', 'Price'], ascending=(False, True))
            posrt2 = pd.concat([posrt2], ignore_index=True)
            i = 0
            while i < len(posrt['Name']):
                if posrt2['Name'][i] != posrt['Name'][0] and posrt2['Name'][i] != posrt['Name'][1] and posrt2['Name'][i] != posrt['Name'][2] and posrt2['Name'][i] != posrt['Name'][3] and posrt2['Name'][i] != posrt['Name'][4] and posrt2['Rating'][i] >= 4.3:
                    medium_panel = tk.Label(stronka1, bg="white", fg="red", width=200,
                                        text="Also recomend for you, but cheep",
                                        font=18, anchor="center")
                    medium_panel.pack()
                    top5 = tk.Label(stronka1, bg="white", fg="black", width=200,
                                    text=posrt2['Name'][i] + ' in price ' + str(
                                        posrt2['Price'][i]) + ' Euro, with medium rating ' + str(
                                        posrt2['Rating'][i]) + ' from ' + str(posrt2['Winery'][i]) + ' winery',
                                    font=10, anchor="center")
                    top5['font'] = f
                    top5.pack()
                    break
                else:
                    i += 1

        helpf = posrt.groupby(by=['Country', 'WineStyle', 'NumberOfRatings'])['Rating'].agg(['mean', 'count'])
        hel = helpf.reset_index()
        hel = hel.loc[hel['count'] >= 5]
        pomocnik = hel.sort_values(by=['mean'], ascending=False)
        pomocnik = pd.concat([pomocnik], ignore_index=True)
        if len(pomocnik['mean'] != 0):
            low_panel = tk.Label(stronka1, bg="white", fg="red", width=200,
                                text="Overal recomendation for you:",
                                font=18, anchor="center")
            low_panel.pack()
            if kraj == 'none':
                if styl == 'none':
                    if czestosc == 'none':
                        top5 = tk.Label(stronka1, bg="white", fg="black", width=200,
                                        text="Our recomendation for your preference is to looking for wines from " + pomocnik['Country'][0] + '. Recomended wine style for you is ' + pomocnik['WineStyle'][0] + '. Recomended popularity of wine for you is ' +pomocnik['NumberOfRatings'][0] + '.',
                                        font=10, anchor="center")
                        top5['font'] = f
                        top5.pack()
                    else:
                        top5 = tk.Label(stronka1, bg="white", fg="black", width=200,
                                        text="Our recomendation for your preference is to looking for wines from " +
                                             pomocnik['Country'][0] + '. Recomended wine style for you is ' +
                                             pomocnik['WineStyle'][0] + '.',
                                        font=10, anchor="center")
                        top5['font'] = f
                        top5.pack()
                else:
                    if czestosc == 'none':
                        top5 = tk.Label(stronka1, bg="white", fg="black", width=200,
                                        text="Our recomendation for your preference is to looking for wines from " + pomocnik['Country'][0] + '. Recomended popularity of wine for you is' +pomocnik['NumberOfRatings'][0] + '.',
                                        font=10, anchor="center")
                        top5['font'] = f
                        top5.pack()
                    else:
                        top5 = tk.Label(stronka1, bg="white", fg="black", width=200,
                                        text="Our recomendation for your preference is to looking for wines from " +
                                             pomocnik['Country'][0] + '.',
                                        font=10, anchor="center")
                        top5['font'] = f
                        top5.pack()
            else:
                if styl == 'none':
                    if czestosc == 'none':
                        top5 = tk.Label(stronka1, bg="white", fg="black", width=200,
                                        text="Our recomendation for your preference is to looking for " +
                                             pomocnik['WineStyle'][0] + ' wines. Recomended popularity of wine for you is ' +
                                             pomocnik['NumberOfRatings'][0] + '.',
                                        font=10, anchor="center")
                        top5['font'] = f
                        top5.pack()
                    else:
                        top5 = tk.Label(stronka1, bg="white", fg="black", width=200,
                                        text="Our recomendation for your preference is to looking for " +
                                             pomocnik['WineStyle'][
                                                 0] + ' wines. ',
                                        font=10, anchor="center")
                        top5['font'] = f
                        top5.pack()
                else:
                    if czestosc == 'none':
                        top5 = tk.Label(stronka1, bg="white", fg="black", width=200,
                                        text="Our recomendation for your preference is to looking for " +pomocnik['NumberOfRatings'][0] + ' wines.',
                                        font=10, anchor="center")
                        top5['font'] = f
                        top5.pack()
                    else:
                        top5 = tk.Label(stronka1, bg="white", fg="black", width=200,
                                        text="We don't have any recomendation for you.",
                                        font=10, anchor="center")
                        top5['font'] = f
                        top5.pack()
    return f


def glowne(wines):
    stronka = tk.Tk()
    stronka.geometry("500x500")
    stronka.title("Wine parameters selection panel")
    gorny_panel = tk.Label(stronka, bg="white", fg="red", width=70, text="Welcome in wine parameters selection panel",
                           font=18, anchor="center")
    gorny_panel.pack()
    country_panel = tk.Label(stronka, fg="black", width=30, text="Choose country ",
                           font=8, anchor="center")
    country_panel.pack()
    k_country = tk.StringVar()
    k_country.set('none')
    klik_country = tk.OptionMenu(stronka, k_country, 'none', 'France', 'Italy', 'Austria', 'New Zealand', 'Chile', 'Australia',
       'South Africa', 'Spain', 'United States', 'Portugal', 'Hungary',
       'Brazil', 'Argentina', 'Romania', 'Germany', 'Greece', 'Mexico',
       'Moldova', 'Switzerland', 'Slovenia', 'Israel', 'Georgia',
       'Lebanon', 'Uruguay', 'Turkey', 'Croatia', 'China', 'Slovakia',
       'Bulgaria', 'Canada', 'Luxembourg', 'Czech Republic',
       'United Kingdom')
    klik_country.pack()
    wine_style_panel = tk.Label(stronka, fg="black", width=30, text="Choose wine style ",
                             font=8, anchor="center")
    wine_style_panel.pack()
    k_winestyle = tk.StringVar()
    k_winestyle.set('none')
    klik_country = tk.OptionMenu(stronka, k_winestyle, 'none', 'red', 'rose', 'sparkling', 'white')
    klik_country.pack()
    nor_style_panel = tk.Label(stronka, fg="black", width=30, text="Choose how popular is your wine ",
                                font=8, anchor="center")
    nor_style_panel.pack()
    k_nor = tk.StringVar()
    k_nor.set('none')
    klik_nor = tk.OptionMenu(stronka, k_nor, 'none', 'Unpopular', 'Semi-popular', 'Popular')
    klik_nor.pack()
    price_panel = tk.Label(stronka, fg="black", width=30, text="Choose maximum wine price ",
                               font=8, anchor="center")
    price_panel.pack()
    price_suwak = tk.Entry(stronka, bg="white", fg="black", width=15, font=8, borderwidth=3,
                           highlightbackground="black", highlightcolor='black')
    price_suwak.insert(0, '0')
    price_suwak.pack()
    age_panel = tk.Label(stronka, fg="black", width=30, text="Choose minimum wine age ",
                           font=8, anchor="center")
    age_panel.pack()
    age_suwak = tk.Scale(stronka, bg="white", fg="black", from_=0, to=63,
                           highlightbackground="black", highlightcolor='black', orient=tk.HORIZONTAL)
    age_suwak.pack()
    age2_panel = tk.Label(stronka, fg="black", width=30, text="Choose maximum wine age ",
                           font=8, anchor="center")
    age2_panel.pack()
    age2_suwak = tk.Scale(stronka, bg="white", fg="black", from_=0, to=63,
                           highlightbackground="black", highlightcolor='black', orient=tk.HORIZONTAL)
    age2_suwak.pack()
    price_suwak.pack()
    enterise = tk.Button(stronka, bg="red", fg="black", width=8, borderwidth=0, highlightcolor="gray",
                         highlightbackground="gray", font=2, anchor="s", text="Enter", command=zatwierdz(k_country, k_winestyle, k_nor, price_suwak, age_suwak, age2_suwak, wines, stronka))
    enterise.pack()
    stronka.mainloop()
    return stronka
glowne(wines)
```
